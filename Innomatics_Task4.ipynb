**Problem 1: Mobile Recharge Validation System**

```python
def validate_recharge(amount):
    valid_plans = [199, 299, 399, 599]
    if amount>=50 and amount in valid_plans:
        return True
    else:
        return False
while True:
    amount=int(input("Enter amount for Recharge:"))
    if validate_recharge(amount):
        print("Recharge Successful!")
        break
    else:
        print("Invalid recharge amount. Please try again.")
```

**Problem 2: Inventory Reorder Alert System**

```python
products = {
    "Rice": 20,
    "Sugar": 10,
    "Oil": 5,
    "Soap": 25
}
def reorder_alert(products):
    for product, quantity in products.items():
        if quantity < 15:
            print(product,":Reorder alert")
        else:
            print(product,":Stock OK")
reorder_alert(products)
    
```

**Problem 3: Student Result Processing System**

```python


def exam_system(marks):
        total=sum(marks)
        avg=total/len(marks)
        print("Average:",avg)
        if avg>=50:
            print("pass")
        else:
            print("fail")
        
n=int(input("Enter number of subjects:"))
marks=[]
for i in range(n):
    mark=int(input("Enter your marks:"))
    marks.append(mark)
    
            
exam_system(marks)
         
```

**Problem 4: Cab Fare Estimator with Retry Option**


```python
def calculate_fare(distance, peak):
    base_fare = 50
    per_km = 12

    fare = base_fare + (per_km * distance)

    if peak == "yes":
        fare = fare * 1.25  

    return fare
while True:
    distance = float(input("Enter distance in km: "))
    peak = input("Is it peak hour? (yes/no): ").lower()
    total_fare = calculate_fare(distance, peak)
    print("Total Cab Fare:", total_fare)
    retry = input("Do you want to calculate again? (yes/no): ").lower()
    if retry != "yes":
        print("Thank you for using Cab Fare Estimator!")
        break
```

**Problem 5: Employee Attendance Eligibility Checker**

```python
attendance = ["P", "P", "A", "P", "P"]
def attendance_checker(attendance):
    count=0
    for i in attendance:
        if i=='P':
           count+=1
    avg=count/len(attendance)*100
    print(avg)

    if avg>=75.0:
        print("Eligible")
    else:
        print("Not Eligible")

attendance_checker(attendance)

    
    
```

**Problem 6: Password Strength Checker**


```python
def check_password(password):

    special = "!@#$%^&*"
    valid = True
    if len(password) < 8:
        print("Password must be at least 8 characters long.")
        valid = False

    digit = False
    special_char = False
    for ch in password:
        if ch.isdigit():
            digit = True
        if ch in special:
            special_char = True

    if digit == False:
        print("Password must contain at least one digit.")
        valid = False

    if special_char == False:
        print("Password must contain at least one special character.")
        valid = False

    return valid

password = input("Enter password: ")

if check_password(password):
    print("Password is valid.")
else:
    print("Please try again.")
```
