**Problem 1: Social Media – Post Engagement Analyzer**

```python
def social_media():
    like_count = 0

    while True:
        like = input("Like the post? (Yes/No): ").lower()

        if like == "yes":
            like_count += 1
        elif like == "no":
            break
        else:
            print("Invalid input")

    if like_count >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    print("Total Likes:", like_count)
    print("Post Status:", status)
social_media()
```

**Problem 3: Agriculture – Rainfall Adequacy Checker**

```python
def analyze_rainfall(rainfall_data, required_level):
    total = 0
    for rain in rainfall_data:
        total += rain
    average = total / len(rainfall_data)
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"
    print("Average Rainfall:", int(average))
    print("Rainfall Status:", status)

rainfall = [70, 75, 80, 65, 70]
required_level = 70
analyze_rainfall(rainfall, required_level)
```

**Problem 6: Agriculture – Premium Crop Price Filter**

```python
def filter_premium_crops(prices):
    premium = []
    for price in prices:
        if price > 2000:
            premium.append(price)
    
    return premium
crop_prices = [1500, 2500, 1800, 3200, 1900]
premium_crops = filter_premium_crops(crop_prices)
print("Premium Crops:", premium_crops)
```

**Problem 10: Smart Electricity Bill Analyzer**

```python
def calculate_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)
    if bill < 500:
        status = "Low Usage"
    elif bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"
    return bill, status
 units = int(input("Enter units consumed: "))
total_bill, usage_status= calculate_bill(units)
print("Total Electricity Bill: ", total_bill)
print("Usage Level:", usage_status)
```
