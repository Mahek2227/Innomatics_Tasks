**Problem Statement 1:Smart Parking Lot Management System**

```python
def parking_system(capacity, logs):
    parked = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT":
            if parked > 0:
                parked -= 1
        if parked > peak_usage:
            peak_usage = parked
    print("Currently Parked Vehicles:", parked)
    if parked > capacity:
        print("Parking Status: Capacity Exceeded")
    else:
        print("Parking Status: Available")

    print("Peak Parking Usage:", peak_usage)

capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

parking_system(capacity, logs)
```

**Problem Statement 2:Online Food Delivery Time Estimator**

```python
def estimate_delivery_time(distance, traffic, weather):
    eta = distance * 3
    if traffic == "Low":
        eta += 5
    elif traffic == "Medium":
        eta += 10
    elif traffic == "High":
        eta += 20

    if weather == "Clear":
        eta += 0
    elif weather == "Cloudy":
        eta += 5
    elif weather == "Rainy":
        eta += 11

    return eta

distance = 8
traffic = "High"
weather = "Rainy"
final_time = estimate_delivery_time(distance, traffic, weather)
print("Estimated Delivery Time:", final_time, "minutes")
```

**Problem Statement 3:Movie Theatre Seat Occupancy Analyzer**

```python
def analyze_seat_occupancy(total_seats, booked_seats):
    booked_count = len(booked_seats)
    occupancy = (booked_count / total_seats) * 100
    print("Occupancy:", int(occupancy), "%")
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 70:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")
    if occupancy >= 90:
        print("Suggestion: Consider opening additional show")

total_seats = 200
booked_seats = [1] * 150 
analyze_seat_occupancy(total_seats, booked_seats)
```

**Problem Statement 4:Cloud Server Load Classification System**

```python
def classify_server_load(cpu_readings):
    total = 0
    for value in cpu_readings:
        total += value
    average = total / len(cpu_readings)
    print("Average CPU Load:", int(average), "%")
    if average < 50:
        status = "Normal"
    elif average <= 80:
        status = "Warning"
    else:
        status = "Critical"
    print("Server Status:", status)
cpu_readings = [45, 60, 70, 85, 90]
classify_server_load(cpu_readings)
```

**Problem Statement 5:Smart Classroom Resource Usage Monitor**


```python
def monitor_resources(resource_usage):
    overused = []
    for resource, hours in resource_usage.items():
        if hours > 8:
            overused.append(resource)
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")

usage = {
    "Projector": 6,
    "AC": 9,
    "Lights": 4
}
monitor_resources(usage)
```

**Problem Statement 6: Online Event Registration Capacity Controller**

```python
def manage_registration(capacity, total_registrations):
    confirmed = 0
    waitlist = 0
    for i in range(total_registrations):
        if confirmed < capacity:
            confirmed += 1
        else:
            waitlist += 1
    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlist)

    if confirmed >= capacity:
        print("Registration Status: Closed")
    else:
        print("Registration Status: Open")

capacity = 100
registrations = 105
manage_registration(capacity, registrations)
```
