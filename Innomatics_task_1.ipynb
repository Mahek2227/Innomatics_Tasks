2.Pass / Fail Analyzer

```python
marks = [45, 78, 90, 33, 60]
count=0
i=0
for students  in marks:
  if students>=50:
     count+=1
  else:
    i+=1
print("total pass students=",count)
print("total fail students=",i)


```

3.Simple Data Cleaner

```python
names = [" Alice ", "bob", " CHARLIE "]
cleaned_names=[]
for i in names:
  cleaned_names.append(i.strip().lower())
print("cleaned list:",cleaned_names)
```

4. Message Length Analyzer

```python
messages = ["Hi", "Welcome to the platform", "OK"]
for i in messages:
  print(f"Message:",i,"Length:",len(i))
  if len(i)>10:
    print("Flagged,Message is too long")

```

5. Error Message Detector

```python
logs = ["INFO", "ERROR", "WARNING", "ERROR"]
count=0
for i in logs:
  if i=="ERROR":
    count+=1
print("Number of Errors=",count)
```
