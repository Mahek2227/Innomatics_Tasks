```python
llm = ChatOpenAI(temperature=0)
```

```python
resume = """
Python, Machine Learning, SQL, NLP project, 1 year internship experience
"""

job_desc = """
Looking for Data Scientist with Python, Machine Learning, Deep Learning, SQL experience
"""
```

```python
extract_prompt = PromptTemplate(
    input_variables=["resume"],
    template="""
Extract:
- Skills
- Tools
- Experience

Resume:
{resume}
"""
)

extract_chain = extract_prompt | llm
extract_output = extract_chain.invoke({"resume": resume})

print("Extraction:\n", extract_output)
```

```python
match_prompt = PromptTemplate(
    input_variables=["resume", "job_desc"],
    template="""
Compare resume with job description.

Give:
- Matching skills
- Missing skills

Resume:
{resume}

Job Description:
{job_desc}
"""
)

match_chain = match_prompt | llm
match_output = match_chain.invoke({
    "resume": resume,
    "job_desc": job_desc
})

print("Matching:\n", match_output)
```

```python
score_prompt = PromptTemplate(
    input_variables=["match"],
    template="""
Give a score from 0 to 100 based on match.

Match:
{match}

Only return number.
"""
)

score_chain = score_prompt | llm
score_output = score_chain.invoke({"match": match_output})

print("Score:\n", score_output)
```

```python
explain_prompt = PromptTemplate(
    input_variables=["score", "match"],
    template="""
Explain why this score was given.

Score:
{score}

Match:
{match}
"""
)

explain_chain = explain_prompt | llm
explain_output = explain_chain.invoke({
    "score": score_output,
    "match": match_output
})

```

```python
print("\n FINAL OUTPUT ")
print("Score:\n", score_output)
print("Explanation:\n", explain_output)
```
