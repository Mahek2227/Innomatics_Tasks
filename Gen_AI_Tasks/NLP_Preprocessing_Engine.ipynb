**Task 1 – Conceptual Understanding**

**1. Difference between "Love" and "love" in NLP**

In Natural Language Processing, text is case sensitive.
"Love" and "love" may be treated as two different tokens if case normalization is not applied.

Example:

Love → 1 occurrence
love → 2 occurrences

To avoid this duplication, most NLP pipelines convert text to lowercase.

**2. What happens if stopwords are not removed?**

Stopwords are common words such as the, is, in, at, and, a.

If they are not removed:

Vocabulary size increases
Models become slower
Important words may get less weight

**3. When removing stopwords is harmful**

Sentiment Analysis

Example:
"I am not happy"

If "not" is removed →
"I am happy" → wrong sentiment.

Question Answering Systems

Example:
"What is the capital of India?"

Removing stopwords may break sentence meaning.

**4. Stemming vs Lemmatization**

Feature	 |  Stemming	     |   Lemmatization

Method	 | Cuts words	  |   Uses dictionary

Accuracy	|Less accurate	|   More accurate

Example	|  running → runn	| running → run

Stemming is faster but rough.
Lemmatization gives proper root words

**Task 2 – Advanced Preprocessing Function**

```python
def preprocess_text(text):

    if not isinstance(text, str) or text.strip() == "":
        return [], ""

    # Lowercase
    text = text.lower()

    # Remove URLs
    text = re.sub(r'http\S+|www\S+', '', text)

    # Remove emails
    text = re.sub(r'\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove short tokens except no/not
    tokens = [word for word in tokens if len(word) > 2 or word in ["no","not"]]

    cleaned_sentence = " ".join(tokens)

    return tokens, cleaned_sentence
```

**Task 3 – Stress Testing**

```python
sentences = [
"Get 100% FREE access now!!!",
"I absolutely looooved this product 😍😍",
"Worst service ever... 0/10",
"Call me at 9876543210",
"This is THE best course!!!",
"Visit https://openai.com now!",
"Nooooo this is baaad!!!",
"OK OK OK I got it",
"Win $$$ now!!! Limited offer!!!",
"I am not happy with this"
]

results = []

for s in sentences:
    tokens, clean = preprocess_text(s)

    print("Original:", s)
    print("Tokens:", tokens)
    print("Clean Sentence:", clean)
    print("-"*40)

    results.append(tokens)
```

**Task 4 – Token Analytics**

```python
for s in sentences:

    tokens, clean = preprocess_text(s)

    if len(tokens) == 0:
        avg_len = 0
    else:
        avg_len = np.mean([len(t) for t in tokens])

    print("Sentence:", s)
    print("Total Tokens:", len(tokens))
    print("Unique Tokens:", len(set(tokens)))
    print("Average Token Length:", round(avg_len,2))
    print("-"*40)
```

**Task 5 – Frequency Analysis**

```python
all_tokens = []

for s in sentences:
    tokens, _ = preprocess_text(s)
    all_tokens.extend(tokens)

freq = Counter(all_tokens)

print("Top 10 Most Frequent Words:")
print(freq.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(freq.most_common()[-5:])
```

**Task 6 – Full NLP Pipeline**

```python
def full_pipeline(text_list):

    all_tokens = []
    clean_sentences = []

    for text in text_list:

        tokens, clean = preprocess_text(text)

        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
print(full_pipeline(sentences))
```
