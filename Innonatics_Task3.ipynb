**Problem Statement 1: Employee Performance Bonus Eligibility**

```python
employees = {
"Ravi": 92,
"Anita": 88,
"Kiran": 92,
"Suresh": 85
}
highest_score=0
highest_names=[]
for name,score in employees.items():
    if score>highest_score:
        highest_score=score
        highest_names=[name]
    elif score==highest_score:
        highest_names.append(name)
print("Top Performers Eligible for Bonus: ",highest_names,"Score:",highest_score)
```

**Problem Statement 2: Search Query Keyword Analysis**

```python
text="Buy mobile phone buy phone online"
words=text.lower().split()
freq={}
for word in words:
    freq[word] = freq.get(word, 0) + 1
result={}

for word, count in freq.items():
    if count>1:
        result[word]=count        
print(result)

        
```

**Problem Statement 4: Email Domain Usage Analysis**

```python
emails = [
"ravi@gmail.com",
"anita@yahoo.com",
"kiran@gmail.com",
"suresh@gmail.com",
"meena@yahoo.com"
]
domain_count={}
for email in emails:
    domain = email.split("@")[1] 
    
    if domain in domain_count:
        domain_count[domain] += 1
    else:
        domain_count[domain] = 1

total_emails = len(emails)
print("Domain Percentage:")
for domain, count in domain_count.items():
    percentage = (count/total_emails) * 100
    print(domain,":",percentage,"%")
```

**Problem Statement 5: Sales Spike Detection**

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
average = sum(sales) / len(sales)
threshold = average * 1.3
for index, value in enumerate(sales, start=1):
    if value > threshold:
        print(f"Day {index}: {value}")
```

**Problem Statement 6: Duplicate User ID Detection**

```python
# Problem Statement 6: Duplicate User ID Detection
# This program identifies duplicate user IDs
# and displays how many times each duplicate appears.

# Input list of user IDs
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]

# Create an empty dictionary to store count of each user ID
user_count = {}

# Loop through each user ID in the list
for user in user_ids:
    if user in user_count:
        user_count[user] += 1   # Increase count if already exists
    else:
        user_count[user] = 1    # Add new user with count 1

# Display duplicate user IDs and their counts
print("Duplicate User IDs:")
for user, count in user_count.items():
    if count > 1:   # Only print duplicates
        print(user, "→", count, "times")
```
