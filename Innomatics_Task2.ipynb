**Problem 1: Unique Words in a Sentence**

```python
sentence="python is easy and python is powerful"
my_list=sentence.split()
print(my_list)
my_set=set(my_list)
print("Unique words count:",len(my_set))
print("Unique words:",my_set)

 
```

**Problem 2: Highest Salary from Employee Data**

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
highest_salary=0
highest_name=''
for name,salary in employees.items():
    if salary > highest_salary:
        highest_salary=salary
        highest_name=name
print("Highest Salary:",highest_name,"-",highest_salary)
    
```

**Problem 3: Find Maximum and Minimum Values**

```python
numbers = [45, 22, 89, 10, 66]
Max=numbers[0]
Min=numbers[0]
for i in numbers:
    if i > Max:
        Max=i
    if i < Min:
        Min=i
print("List:",numbers)
print("Max:",Max)
print("Min:",Min)
    
        
    
    
```

**Problem 4: Count Products Above a Price Threshold**

```python
prices = [450, 1200, 899, 1500, 300]
count=0
for i in prices:
    if i >1000:
        count+=1
print("Products above 1000:",count)
```

**Problem 5: Calculate Attendance Percentage**


```python
attendance = ["P", "P", "A", "P", "P"]
count=0
for i in attendance:
    if i=="P":
        count+=1
average=len(attendance)
total_present=(count/average)*100
print("Attendance Percentage:",total_present)




```

**Problem 7: Count Character Frequency**

```python
text="pythonp"
freq={}
for ch in text:
    if ch in freq:
        freq[ch] +=1
    else:
        freq[ch]=1
print(freq)
    
    
```

**Problem 9: Check if a Key Exists in a Dictionary**


```python
employees={
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}
user_input = input("Enter Employee Name:").capitalize()
if user_input in employees:
    print("Employee Exist")
else:
    print("Employee doesn't Exist")
```
